```python
import os
import sys
import pickle

sys.path.insert(0, '..')
sys.path.insert(0, '../..')
sys.path.insert(0, '../../..')
sys.path.insert(0, '../../../..')
sys.path.insert(0, '../../../../..')

from load_data.load_conformal_data import load_results

from conformance_checking.conformance import ConformanceChecking

from conformance_checking.conformance_results import ConformanceResults
```

```python
# Compute the fitness scores for all probabilistic sffix samples for all cases in the validation dataset:

path_to_be_model = '../../../../process_models/Helpdesk.bpmn'

path_d_con_results = '../../../../../../data/Helpdesk/proact_conf_check/eval_validation_set/conformal_val'

# MC Suffix Sampling Results
results = load_results(path=path_d_con_results)

# Conformance Checking object:
cc = ConformanceChecking(path_process_model=path_to_be_model)

# Evaluated the probabilistic suffix samples using conformance checking:
cfr = ConformanceResults(log_name='Helpdesk', data=results, conformance_object=cc)

fitness_dict = cfr.fitness_scores_calibration()

# Save results to .pkl in conformal_val directory
output_dir   = '../../../../../../data/Helpdesk/proact_conf_check/offline_analysis'

os.makedirs(output_dir, exist_ok=True)  # Ensure the directory exists

# Package results into a dictionary
results = fitness_dict
filename = os.path.join(output_dir, f'fitness_results.pkl')
with open(filename, 'wb') as f:
    pickle.dump(results, f)
print(f"Saved {len(results)} entries to {filename}")

print("Finished conformance checking!")
```
