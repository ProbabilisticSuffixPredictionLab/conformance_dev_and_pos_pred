```python
import os
import sys
import pickle

sys.path.insert(0, '..')
sys.path.insert(0, '../..')
sys.path.insert(0, '../../..')
sys.path.insert(0, '../../../..')
sys.path.insert(0, '../../../../..')

from load_data.load_conformal_data import load_results
from conformance_checking.conformance import ConformanceChecking
from conformance_checking.conformance_results import ConformanceResults
```

```python
# Load model and validation set and determine output directory:

# Sepsis:
log_name = "Sepsis"
path_to_be_model = "../../../../data/process_models/Sepsis.bpmn"
path_calibration_samples_results = "../../../../../../data/Sepsis/proact_conf_check/eval_validation_set/"
output_dir   = "../../../../../../data/Sepsis/proact_conf_check/offline_analysis/calibration_conformance_results"

# Helpdesk
```

```python
# Compute the fitness scores for all probabilistic sffix samples for all cases in the validation dataset:
# MC Suffix Sampling Results
results = load_results(path=path_calibration_samples_results)

# Conformance Checking object:
cc = ConformanceChecking(path_process_model=path_to_be_model)
# Evaluate the probabilistic suffix samples using conformance checking:
cfr = ConformanceResults(log_name=log_name, data=results, conformance_object=cc)
fitness_dict = cfr.conformance_results(target_workers=128)
# Testing purpose 
# test = cfr.quick_fitness()

# Think about iterating through the dict and storing every 50 in .pkl

os.makedirs(output_dir, exist_ok=True)  # Ensure the directory exists
# Package results into a dictionary
results = fitness_dict
filename = os.path.join(output_dir, f'conformance_calibration.pkl')
with open(filename, 'wb') as f:
     pickle.dump(results, f)
print(f"Saved {len(results)} entries to {filename}")
print("Finished conformance checking for calibration data!")
```
