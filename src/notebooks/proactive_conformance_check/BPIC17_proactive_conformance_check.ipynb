```python
cc = ConformanceChecking(path_process_model=path_pm, path_prob_suffix_pred=path_sampled_cases)
```

```python
d = cc.get_case_ids()

print("Pickle Results: ", d)
print("Amount prefixes in pkl: ", len(list(d)))

app_names = set(key[0] for key in d)
app_names_list = list(app_names)

print(app_names_list)
```

```python
list_predicted_sample_case_alignment, list_most_likely_case_alignment , list_target_case_alignment = cc.conformance_of_sampled_suffixes(case_concept_name=app_names_list[1])
```

```python
# Fitness score visualizations:
for i, predicted_sample_case_alignment in enumerate(list_predicted_sample_case_alignment):

    sampled_suffixes_case = [res['sampled case'] for res in predicted_sample_case_alignment]
    sampled_suffixes_fitness = [res['fitness'] for res in predicted_sample_case_alignment]
    sampled_suffixes_case = [res['cost'] for res in predicted_sample_case_alignment]
    
    mean_sampled_suffixes_fitness = sum(sampled_suffixes_fitness)/ len(sampled_suffixes_fitness)
    
    # Get the mean fitness of all sampled cases
    print("Mean of all Sampled Predicted Case Fitness: ", mean_sampled_suffixes_fitness)
    
    most_likely_case = list_most_likely_case_alignment[i]['most likely case']
    most_likely_align = list_most_likely_case_alignment[i]['alignment']
    most_likely_fitness = list_most_likely_case_alignment[i]['fitness']
    most_likely_cost = list_most_likely_case_alignment[i]['cost']
    
    print("Most-likely Predicted Case: ", most_likely_case)
    print("Most-likely Predicted Case Alignment: ", most_likely_align)
    print("Most-likely Predicted Case Fitness: ", most_likely_fitness)
    
    target_case = list_target_case_alignment[i]['target case']
    target_align = list_target_case_alignment[i]['alignment']
    target_fitness = list_target_case_alignment[i]['fitness']
    target_cost = list_target_case_alignment[i]['cost']
    
    print("Target Case: ", target_case)
    print("Target Case Alignment: ", target_align)
    print("Target Case Fitness: ", target_fitness)
    
    plt.hist(sampled_suffixes_fitness, bins=50, edgecolor='black')
    plt.xlabel('Value')
    
    plt.axvline(x=most_likely_fitness, color='orange', linestyle='--', linewidth=2)
    
    plt.axvline(x=target_fitness, color='green', linestyle='--', linewidth=2)
    
    plt.ylabel('Frequency')
    plt.title('Histogram of Float Values')
    
    plt.grid(True)
    plt.show()
```
